```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus=["i love nlp","i love python"]
Vectorizer=TfidfVectorizer()
X=Vectorizer.fit_transform(corpus)

print(Vectorizer.get_feature_names_out())
print(X.toarray())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

corpus=["I love NLP","I love Python"]
vectorizer=CountVectorizer()
X=vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

```python
import gensim
from gensim import corpora

docs=[
    "i love pizza and pasta"
    "cars and engines are amazing"
    "cheese pizza is the best"
    "i drive my car every day"
]
texts=[doc.lower().split() for doc in docs]
dictionary=corpora.Dictionary(texts)
corpus=[dictionary.doc2bow(text) for text in texts]

lda_model = gensim.models.LdaModel(corpus,num_topics=2,passes=10)
topics=lda_model.print_topics()
for topic in topics:
    print(topic)

```
